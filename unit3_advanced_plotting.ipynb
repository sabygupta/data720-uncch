```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
from matplotlib.gridspec import GridSpec

from PIL import Image

with Image.open("cells.png") as img:
    img_array = np.array(img)
```

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.exp(x)

fig, ax1 = plt.subplots()
ax1.plot(x, y1, c="blue")
ax1.set_ylabel("Sin(x)", color="blue")
ax1.tick_params(axis="y", labelcolor="blue")

ax2 = ax1.twinx()
ax2.plot(x, y2, c="red")
ax2.set_ylabel("Exp(x)", color="red")
ax2.tick_params(axis="y", labelcolor="red")

plt.show()
 
```

```python
fig, ax = plt.subplots()

ax.plot(x, y1)
ax.set_title("Main plot")

axins = inset_axes(ax, width="30%", height="30%", loc="lower right")
axins.plot(x, y2, c="red")
plt.show()
```

```python
y3 = np.cos(x)

fig = plt.figure(figsize=(8,6))
gs = GridSpec(2, 2, figure=fig)
ax1 = fig.add_subplot(gs[0, :])
ax1.plot(x, y1)

ax2 = fig.add_subplot(gs[1, 0])
ax2.plot(x, y2)

ax3 = fig.add_subplot(gs[1, 1])
ax3.plot(x, y3)

plt.show()
```

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")
ax.plot_surface(X, Y, Z, cmap="viridis")
plt.show()
```
